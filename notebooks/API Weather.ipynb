```python
import pandas as pd
import requests
```

```python
#https://open-meteo.com/en/docs/historical-weather-api

url_weather = "https://archive-api.open-meteo.com/v1/archive?latitude=41.39&longitude=2.16&start_date=2021-01-01&end_date=2023-04-28&daily=temperature_2m_mean,rain_sum&timezone=auto"
res = requests.get(url_weather).json()
```

```python
print(res)
```

```python
df = pd.DataFrame(res['daily'])
df.dropna(subset=['temperature_2m_mean'], inplace=True) 
df = df.rename(columns={'time': 'Date', 'temperature_2m_mean': 'Temp_mean', 'rain_sum': 'Rain'})
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce') 
df['Rain'] = df['Rain'] > 0.0
df
```

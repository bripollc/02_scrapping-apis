```python
import pandas as pd
import requests
```

```python
#https://open-meteo.com/en/docs/historical-weather-api

url_weather = "https://archive-api.open-meteo.com/v1/archive?latitude=41.39&longitude=2.16&start_date=2021-12-01&end_date=2023-04-28&daily=temperature_2m_mean,rain_sum&timezone=auto"
res = requests.get(url_weather).json()
res
```

```python
df = pd.DataFrame(res['daily'])
df.dropna(subset=['temperature_2m_mean'], inplace=True) 
```
